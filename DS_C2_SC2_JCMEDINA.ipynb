```python
import pandas as pd
```

##Acción A

Extraer la información del archivo.

```python
vacunas = pd.read_csv ("sample_data/country_vaccinations.csv")
```

##Acción B

Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
vacunas['date'] = pd.to_datetime(vacunas['date'])
vacunas.info()
```

##Acción C

Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
df = vacunas.copy(deep=True)
df['vaccines'] = df['vaccines'].str.split(',')
df_exploded = df.explode('vaccines')
df_exploded.groupby('vaccines')['people_vaccinated'].sum().reset_index()
```

##Acción D


Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
print("{:10,.0f}".format(vacunas['people_vaccinated'].sum()))
```

##Acción E

Calcular el promedio de vacunas aplicadas por país.

```python
vacPais = vacunas[['country','people_vaccinated']].groupby('country').mean()
vacPais['people_vaccinated'] = vacPais['people_vaccinated'].astype(int)
vacPais
```

##Acción F

Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
print("{:10,.0f}".format(vacunas[vacunas['date']=='29/01/2021']['people_vaccinated'].sum()))
```

##Acción G

Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

```python
conDiferencias = vacunas.copy(deep=True)
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias
```

##Acción H

Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
max(vacunas['date']) - min(vacunas['date'])
```

##Acción I

Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
conCantidad = vacunas.copy(deep=True)
conCantidad['canVac'] = conCantidad['vaccines'].str.split(",").str.len()
conCantidad
```

##Acción J

Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = vacunas.query('date<="20/12/2020"')
antes20
```

##Acción K

Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = vacunas[vacunas['vaccines'].str.contains('Pfizer')]
pfizer
```

##Acción L

Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
with pd.ExcelWriter("/content/resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
